```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from ipywidgets import Text
%matplotlib widget

def f(x):
    print(x)
    a = np.arange(1,x,1)
    plt.clf()
    plt.plot(a)
    return


w = widgets.IntSlider(
    value=7,
    min=0,
    max=10,
    step=1,
    description='Test:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)




widgets.interact(f,x = w)





```
